```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras.utils.np_utils import to_categorical
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('mitbih_train.csv', header = None)
```

```python
test = pd.read_csv('mitbih_test.csv', header = None)
```

```python
train.head()
```

```python
train.describe()
```

```python
train[187]=train[187].astype(int)
train[187].value_counts()
#0 - normal beats
#1 - supraventicular ectopic beats
#2 - ventricular ectopic beats
#3 - fusion beats
#4 - Unclassified beats
```

```python
# we use resample technique to decrease this huge difference in the balance of classes
from sklearn.utils import resample
```

```python
t0 = train[train[187] == 0]
t1 = train[train[187] == 1]
t2 = train[train[187] == 2]
t3 = train[train[187] == 3]
t4 = train[train[187] == 4]

t0_resampled = resample(t0, replace = True, n_samples = 20000, random_state = 42)
t1_resampled = resample(t1, replace = True, n_samples = 20000, random_state = 43)
t2_resampled = resample(t2, replace = True, n_samples = 20000, random_state = 44)
t3_resampled = resample(t3, replace = True, n_samples = 20000, random_state = 45)
t4_resampled = resample(t4, replace = True, n_samples = 20000, random_state = 46)

train = pd.concat([t0_resampled,t1_resampled,t2_resampled,t3_resampled,t4_resampled])
```

```python
train[187].value_counts()
#resampled data
```

```python
train0=train[187]
test0=test[187]
y_train=to_categorical(train0)
y_test=to_categorical(test0)
```

```python
X_train=train.iloc[:,:186].values
X_test=test.iloc[:,:186].values
X_train = X_train.reshape(len(X_train), X_train.shape[1],1)
X_test = X_test.reshape(len(X_test), X_test.shape[1],1)
```

```python
import tensorflow as tf
```

```python
data_shape = X_train.shape[1],1
```

# Initialising the CNN

```python
cnn = tf.keras.models.Sequential()
```

```python
#convolution and pooling layer 1
cnn.add(tf.keras.layers.Convolution1D(filters = 64, kernel_size = 3, activation = 'relu', input_shape = data_shape))
cnn.add(tf.keras.layers.MaxPooling1D(pool_size = 3, strides =2, padding = "same"))
```

```python
#convolution and pooling layer 2
cnn.add(tf.keras.layers.Convolution1D(filters = 64, kernel_size = 3, activation = 'relu', input_shape = data_shape))
cnn.add(tf.keras.layers.MaxPooling1D(pool_size = 3, strides =2, padding = "same"))
```

```python
#convolution and pooling layer 3
cnn.add(tf.keras.layers.Convolution1D(filters = 64, kernel_size = 3, activation = 'relu', input_shape = data_shape))
cnn.add(tf.keras.layers.MaxPooling1D(pool_size = 3, strides =2, padding = "same"))
```

```python
#flattening
cnn.add(tf.keras.layers.Flatten())
```

```python
#full connection
cnn.add(tf.keras.layers.Dense(units = 64, activation ='relu'))
```

```python
#full connection
cnn.add(tf.keras.layers.Dense(units = 32, activation ='relu'))
```

```python
#output layer
cnn.add(tf.keras.layers.Dense(units = 5, activation ='softmax'))
```

```python
#compiling
cnn.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics= ['accuracy'])
```

```python
#training
cnn.fit(X_train, y_train, epochs = 40, batch_size = 32, validation_data = (X_test,y_test))
```

```python
scores = cnn.evaluate((X_test),y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores[1]*100))
```

```python
y_pred = cnn.predict(X_test)
```

```python
y_pred
```

```python
prediction = []
for i in range(0,21892):
    if y_pred[i][0] > 0.7:
        prediction.append('normal')
    elif y_pred[i][1] > 0.7:
        prediction.append('supraventricular ectopic beats')
    elif y_pred[i][2] > 0.7:
        prediction.append('Ventricular ectopic beats')
    elif y_pred[i][3] > 0.7:
        prediction.append('Fusion beats')
    elif y_pred[i][4] > 0.7:
        prediction.append('unclassified beats')
```
